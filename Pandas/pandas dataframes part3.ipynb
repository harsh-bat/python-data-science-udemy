```python
import numpy as np
import pandas as pd
```

```python
outside = ['G1','G1','G1','G1','G2','G2','G2']
inside = [1,2,3,4,1,2,3]
hier_index = list(zip(outside, inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
pd.DataFrame(np.random.randn(7,3),hier_index,['A','B','C'])
```

```python
super_out=['A1','A1','A1','A1','A1','A1','A1']
outside = ['G1','G1','G1','G1','G2','G2','G2']
inside = [1,2,3,4,1,2,3]
hier_index = list(zip(super_out,outside, inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
pd.DataFrame(np.random.randn(7,3),hier_index,['A','B','C'])
```
