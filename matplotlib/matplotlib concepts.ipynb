```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# let's you see in jupyter notebook
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show() when not using notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,4)
plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x,y,'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0,0,1,1])
axis2 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
axis3 = fig.add_axes([0.2,0.2,0.4,0.4])
axis1.plot(x,y**4,'b')
axis2.plot(y,x**2,'r')
axis3.plot(x**6,y**30,'g')
```

Part 2

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x**2,y)
axes[1][0].plot(x,y**2)
axes[1][1].plot(x**100,y)
plt.tight_layout()
```

## Figure size and dpi

```python
fig = plt.figure(figsize=(2,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1,figsize=(10,10))
axes[0].plot(x**0.5,y)
axes[1].plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = x**2
```

```python
fig = plt.figure(dpi=125)
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(y,x,label='one')
ax1.plot(x,y,label='two')
ax1.legend(loc=4)
```

Part 3

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
# ax1.plot(x,y,color='red')
# default linewidth=1
ax1.plot(y,x,color='#00FF00', linewidth=4, alpha=0.6)
ax1.plot(x*4,y,color='#FFFF00', lw=4, alpha=0.9)
ax1.plot(x**2,y,color='#000F00', lw=1,linestyle='-.')
ax1.plot(x,y,color='red', lw=1,linestyle='steps')
```

```python
fig1 = plt.figure()
x1 = np.linspace(-10,10,10)
y1 = x1**3
ax2 = fig1.add_axes([0,0,1,0.5])
ax2.plot(x1,y1,marker='o', markersize=20, markerfacecolor='red', markeredgecolor='green', markeredgewidth=5)
```
